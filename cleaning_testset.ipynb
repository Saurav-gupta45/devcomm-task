```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/sauravgupta/Downloads/jc-dev-comm-recruitment-task/test.csv')
```

```python
df_copy3 = df.copy()
```

```python
df_copy3['Age'] = df_copy3['Age'].fillna(df_copy3['Age'].median())
```

```python
df_copy3['Age'].unique()
```

```python
num_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in num_cols:
    df_copy3[col] = df_copy3[col].fillna(df_copy3[col].median())
```

```python
df_copy3['TotalSpend'] = df_copy3[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].sum(axis=1)
```

```python
df_copy3.head()
```

```python
df_copy3['Cabin'] = df_copy3['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
col = ['HomePlanet','CryoSleep','Destination','VIP']
for col_no in col:
    df_copy3[col_no] = df_copy3[col_no].fillna(df_copy3[col_no].mode()[0])
```

```python
df_copy3['cabin_no1'] = df_copy3['Cabin'].str.split('/').str[0]
df_copy3['cabin_no2'] = df_copy3['Cabin'].str.split('/').str[1]
df_copy3['cabin_no3'] = df_copy3['Cabin'].str.split('/').str[2]
```

```python
df_copy3['cabin_no2'] = df_copy3['cabin_no2'].astype(int)
```

```python
df_copy3['Id'] = df_copy3['PassengerId'].str.split('_').str[0]
```

```python
no_of_person = df_copy3['Id'].value_counts()
df_copy3['no_of_person'] = df_copy3['Id'].map(no_of_person)
```

```python
df_copy3['IsAlone'] = df_copy3['no_of_person'] == 1
```

```python
df_copy3.drop(['Name', 'Cabin', 'PassengerId', 'Id'], axis=1, inplace=True)
```

```python
df_copy3['CryoSleep'] = df_copy3['CryoSleep'].map({True:1 , False:0})
```

```python
df_copy3['VIP'] = df_copy3['VIP'].map({True:1 , False:0})
df_copy3['IsAlone'] = df_copy3['IsAlone'].map({True:1 , False:0})
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['HomePlanet', 'Destination', 'cabin_no1','cabin_no3','AgeGroup']
le = LabelEncoder()
for col in label_cols:
    df_copy3[col] = le.fit_transform(df_copy3[col])
```

```python
df_copy3['NoSpend'] = (df_copy3['TotalSpend'] == 0).astype(int)
```

```python
df_copy3['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'])
```

```python
df_copy3.to_csv('clean_test_data.csv',index=False)
```
